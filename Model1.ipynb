```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , classification_report
```

```python
df = pd.read_csv('spam.csv',encoding='latin-1')
df.head()
df.columns
```

```python
if "label" in df.columns and "message" in df.columns:
  df = df[["label" , "message"]]
elif "v1" in df.columns and "v2" in df.columns:
  df = df[["v1","v2"]]
  df.columns = ["label" , "message"]
else:
  raise ValueError("Unknown column format.Check your CSV")
```

```python
df["label"] = df["label"].str.lower().map({
    "ham":0,
    "spam":1
})

df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train , y_test = train_test_split(
    df["message"],
    df["label"],
    test_size=0.2,
    random_state=42,
    stratify=df["label"] #important
)
```

```python
from numpy import vectorize
from sklearn.feature_extraction.text import TfidfVectorizer

vectorize = TfidfVectorizer(
    stop_words="english",
    ngram_range=(1,2),
    min_df=2
)

X_train_vec = vectorize.fit_transform(X_train)
X_test_vec = vectorize.transform(X_test)
```

```python
from sklearn .naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

y_pred = model.predict(X_test_vec)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
def predict_sms(text):
  vec = vectorize.transform([text])
  return "Spam" if model.predict(vec)[0] == 1 else "Ham"

predict_sms("no off tomorrow")
```

```python
import joblib

joblib.dump(model , "spam_model.pkl")
joblib.dump(vectorize,"tfidf_vectorizer.pkl")
```
